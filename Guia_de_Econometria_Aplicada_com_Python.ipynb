### Nomeie as variáveis de forma inteligente:

#### Métodos:

1. Snake Case: É a convenção mais comum em Python, onde as palavras são separadas por underscores (_).
  * Exemplo: nome_completo, idade_usuario
2. Camel Case: Neste estilo, cada palavra, exceto a primeira, é capitalizada e não há espaços ou underscores entre elas.
  * Exemplo: nomeCompleto, idadeUsuario
3. Pascal Case: Semelhante ao Camel Case, mas com a primeira letra maiúscula.
  * Exemplo: NomeCompleto, IdadeUsuario

* *Obs: evite acentos e símbolos/letras que não tem no Inglês, como ç, á...*

## Tipos de dados

* Números Inteiros (`int`): Representam números inteiros, como -3, 0, 42.

Para a regressão, será estimado o seguinte modelo econométrico usando o MQO

### <center>**Modelo:**

##<center> ln(wage) = β0 + β1 ln(educ) + β2 ln(exper) + β3 ln(expersq) + β4 ln(sibs) + u </center>

**OBS.:**

- wage = Salário médio por hora
- as variáveis serao transformadas em log usando o pacote `numpy`.

#### Como calcular o Fator de Inflacão da Variância (FIV ou VIF)?

#### Presentinho: fórmula pronta para um vif tabelado

```python
# criando def para o VIF para fazer a tabela mais facilmente
from statsmodels.stats.outliers_influence import variance_inflation_factor

def vif(x):
    viftab = []
    for i, col in enumerate(x.columns):
        vif = variance_inflation_factor(x.values, i)
        viftab.append((col, vif))
    viftab = pd.DataFrame(viftab)
    #viftab = viftab.drop(0, axis=0) # retirando a coluna do intercepto 'const' se necessario
    viftab = viftab.rename(columns={0: 'Variável', 1: 'VIF'})
    vif_media = viftab['VIF'].mean()
    df_vif_med = pd.DataFrame({'Variável': 'Média', 'VIF': vif_media}, index=[0])
    viftab = pd.concat([viftab, df_vif_med], ignore_index=True)
    x.viftab = viftab
    return viftab
```

```python
# Rodando o VIF
vif(x_sem_log)
```

##### Analisando por Média Movel

$$\hat{y}_{n+1} = \frac{1}{m} \sum_{i=n-m+1}^t y_i$$

Neste método, cada ponto da nova série modelada é calculado como sendo a média dos valores das $m$ observações anteriores. Todas as observações são ponderadas pelo mesmo
valor $(\frac{1}{m})$.

O número de pontos conhecidos da série é representado pelo parâmetro $n$, enquanto $m$ representa a ordem do modelo de média móvel.

Vão ser aplicadas as médias móveis de 3, 6 e 9 períodos.


```python
# Adicionando uma média móvel - 6 períodos
df_acoes_mm1 = df_acoes.copy()
df_acoes_mm1['ABEV MM '+ str(6)] = df_acoes_mm1['ABEV3.SA'].rolling(6).mean()
df_acoes_mm1['PETRO MM '+ str(6)] = df_acoes_mm1['PETR4.SA'].rolling(6).mean()
df_acoes_mm1
```
